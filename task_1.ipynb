### 2. Create a random 100000x1 array from the same exponential distribution and visualize the distribution histogram along with uniform and normal distributions.

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(np.random.exponential(size=(100000, 1)), density=True, bins=100, histtype="step", color="blue", label="exponential")
plt.hist(np.random.randn(100000), density=True, bins=100, histtype="step", color="red", label="randn")
plt.hist(np.random.uniform(size=(10000, 1)), density=True, bins=100, histtype="step", color="green", label="uniform")
plt.axis([-2.5, 2.5, 0, 1.5])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

### 3. Using matplotlib, plot the Z = X^2 + Y^2 in a 3D plot

```python
import matplotlib
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```

### 4. Calculate the pearson standard coefficient and spearman rank correlation for the [HP, Attack, Defense, Sp. Ark, Sp. Def, Speed]

```python
import pandas as pd
import seaborn as sns

```

```python
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
df.head()
```

```python
stats_df = df.drop(['Name', 'Type 1', 'Type 2','Total', 'Stage', 'Legendary'], axis=1)
```

```python
stats_df.head()
```

```python
pearson_corr = stats_df.corr(method='pearson')
pearson_corr
```

```python
spearman_corr = stats_df.corr(method='spearman')
spearman_corr
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Pearson Correlation Coefficient')

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Spearman Correlation Coefficient')

plt.tight_layout()
plt.show()

```
