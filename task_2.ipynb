### 1. Create a new dataframe by filtering all the columns [i.e., features] except N_total, N_bulk, N_surface and R_avg columns.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Au_nanoparticle_dataset.csv', index_col=0, encoding='latin')
df.head()
```

```python
refined_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

### 2. Display the first 20 samples of this dataframe.

```python
refined_df.head(20)
```

### 3. Calculate the mean, standard deviation and quartile values for each of the above 4 features.

```python
refined_df.describe().loc[['mean', 'std', '25%', '50%', '75%']]
```

### 4. Plot the histogram of each of these features in a 1x4 layout.


```python
plt.figure(figsize=(20, 5))

plt.subplot(1, 4, 1)
plt.hist(refined_df['N_total'], bins=100, color='blue', alpha=0.7)
plt.title('N_total Distribution')
plt.xlabel('N_total')
plt.ylabel('Frequency')

plt.subplot(1, 4, 2)
plt.hist(refined_df['N_bulk'], bins=100, color='red', alpha=0.7)
plt.title('N_bulk Distribution')
plt.xlabel('N_bulk')

plt.subplot(1, 4, 3)
plt.hist(refined_df['N_surface'], bins=100, color='orange', alpha=0.7)
plt.title('N_surface Distribution')
plt.xlabel('N_surface')

plt.subplot(1, 4, 4)
plt.hist(refined_df['R_avg'], bins=100, color='purple', alpha=0.7)
plt.title('R_avg Distribution')
plt.xlabel('R_avg')


plt.tight_layout()
plt.show()
```

### 5. Visualize the scatter plots and histograms of this dataframe using the pairplot functionality of seaborn library.

```python
sns.pairplot(refined_df)
plt.show()
```

### 6. Plot the following

#### a. Plots on the diagonal contains the histogram of each feature along with the kernal density estimation plot.

#### b. Plots on the lower half [g.map_lower] contains the bivariate kernal density estimation plot.

```python
g = sns.PairGrid(refined_df)

g.map_upper(sns.kdeplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)

plt.show()
```
